```python
import wandb
from PIL import Image

# Initialize wandb
wandb.init(project="CPEN455HW", entity="idil-bil-ubc")

# Path to the sample directory
sample_dir = "/kaggle/working/CPEN455HW-2024W2/samples"

# List all image files in the sample directory that start with "Class"
image_files = [f for f in os.listdir(sample_dir) if f.startswith('Class') and f.endswith('.png')]

# Organize images by class
class_images = {}
for image in image_files:
    class_name = image.split('_')[0]  # Extracts "Class2" from "Class2_image_10.png"
    if class_name not in class_images:
        class_images[class_name] = []
    class_images[class_name].append(image)

# Log images for each class
for class_name, images in class_images.items():
    wandb.log({
        f"{class_name}_images_grid": [wandb.Image(os.path.join(sample_dir, image)) for image in images]
    })
```
